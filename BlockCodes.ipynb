```python
from encoder.encoder import Encoder
import pandas as pd
import matplotlib.pyplot as plt
```

```python

encoder = Encoder()
character_probabilities = {"A": 7 / 8, "B": 1 / 8}    

def fill_columns(index):
    codes = encoder.get_blocks(character_probabilities, index)
    entropy = encoder.calculate_entropy(codes)
    shannon_average = encoder.calculate_average_length(codes, encoder.get_shannon_code(codes))
    huffman_average = encoder.calculate_average_length(codes, encoder.get_huffman_code(codes))

    return pd.Series({
        'Entropy': entropy,
        'Shannon Average Length': shannon_average,
        'Huffman Average Length': huffman_average

    })
```

```python

df = pd.DataFrame(index=range(1, 11), columns=['Entropy', 'Shannon Average Length', 'Huffman Average Length'])
df = df.apply(lambda row: fill_columns(row.name), axis=1)


df.plot(y=['Entropy', 'Shannon Average Length', 'Huffman Average Length'], marker='o', figsize=(10, 6))

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Plot of Entropy, Shannon Average Length, and Huffman Average Length')


plt.show()
```
